```python
from bs4 import BeautifulSoup as bs
import requests
import lxml
from datetime import date
```

```python
web1='https://www.mining.com/'
web2='https://www.renewableenergyworld.com/#gref'
web3='https://www.miningweekly.com/page/graphite'
web4='https://www.miningweekly.com/page/lithium'
web5='https://www.miningweekly.com/page/rare-earth-minerals'
web6='https://www.miningweekly.com/page/vanadium'
web7='https://www.theassay.com/news-2/'
web8='https://resourceworld.com/category/resources/cobalt/'
web9='https://resourceworld.com/category/resources/graphite/'
web10='https://resourceworld.com/category/resources/lithium/'
web11='https://resourceworld.com/category/resources/nickel/'
web12='https://resourceworld.com/category/resources/vanadium/'
```

```python
html_text=requests.get(web1).text
soup=bs(html_text,'lxml')
articles=soup.find_all('div',class_='col-lg-8')
for article in articles:
        article_link=article.a['href']
        article_title=article.h3.text
        published_date=article.find('div',class_='post-meta').text

        print(f'''
Article Title:{article_title}\n
Article Link: {article_link}\n
Published_date: {published_date}\n
''')

        print('')

```
